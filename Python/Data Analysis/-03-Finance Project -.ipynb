** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
bank_stocks["BAC"].head()
```

```python
#for every ticker in the list of tickers
for tick in tickers:
    #create a new column return['name of ticker' + return]
    #fill that column with the percent change of:
    #the close column of that same ticker column in bank_stocks dataframe
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
bank_stocks['BAC'].head()
```

```python
#Average close price for Bank of America over 30 days
bank_stocks['2008-01-01':'2009-01-01']['BAC']["Close"].rolling(window=30,min_periods=1).mean().plot(figsize=(12,6),label='30 Day Average',color='blue')

#Overall close price for Bank of America
bank_stocks['2008-01-01':'2009-01-01']['BAC']["Close"].plot(label='BAC Close Price',color='green')
plt.legend()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
temp = bank_stocks.xs(key="Close",axis=1,level="Stock Info")
#temp.corr()
sns.heatmap(temp.corr(),cmap='coolwarm',annot=True)
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(temp.corr(),cmap='coolwarm',annot=True)
```

```python
#creating a heatmap using plotly/cufflinks
temp.iplot(kind='heatmap',colorscale='rdylbu')
```
